```python
import pymongo 
```

```python
# connection establishment
client = pymongo.MongoClient('mongodb://localhost:27017')
print('connect')
```

```python
# database creation
```

```python

mydb = client['connnection']
print('done')
```

```python
# create the table
```

```python
mycol = mydb['employees']
print('done')
```

```python
# inserting on value
```

```python
query = {'id':1,'name':'yash','salary':25000}
mycol.insert_one(query)
print('done')
```

```python
# insert many values
```

```python
query = [
    {'id':2,'name':'rohit','salary':15000},
    {'id':3,'name':'rishab','salary':16000},
    {'id':4,'name':'piyush','salary':20000},
    {'id':5,'name':'denial','salary':26000},
    {'id':6,'name':'sanjay','salary':35000},
    {'id':7,'name':'sandeep','salary':40000}
]
mycol.insert_many(query)
print('done')
```

```python
# fatching out data
```

```python
for i in mycol.find({},{'_id':0}):
    print(i)
```

```python
# find data
```

```python
name = input('enter the name: ')
for i in mycol.find({'name':{'$regex':name,'$options':'i'}},{'_id':0}):
    print(i)
```

```python
# updating data
```

```python
Id = int(input('enter the id: '))
name = input('enter the name: ')
salary = int(input('enter the salary: '))
q1 = {'id':Id}
q2 = {'$set':{'name':name,'salary':salary}}
mycol.update_one(q1,q2)
for i in mycol.find({},{'_id':0}):
    print(i)
```
